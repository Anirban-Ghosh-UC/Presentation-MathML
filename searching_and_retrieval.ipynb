```python
import re
import numpy as np
import pandas as pd
with open("./2.txt", "r") as file_2: 
    final_data=file_2.readlines() 
    final_data=[re.sub('\n','',i) for i in final_data]  
    final_data=final_data[1:]
    post_list=[]
    formula_list=[]
    formula_id_list=[]
    sim_list=[]
    relevance_list=[]
    query_id=input("enter the query id: ")
    numList = list(int(num) for num in input("Enter the list numbers separated by space: ").strip().split())
    for line in final_data:
        sim_count,dissim_count=0,0
        formula_id,post,formula=line.split('\t')
        formula=formula.strip().split()
        formula=[int(i) for i in formula]
#         for num in numList:
#             if num in formula:
#                 count+=1
        for i in range(len(numList)):
            if(numList[i]==1 and formula[i]==1):
                sim_count+=1
            elif((numList[i]==1 and formula[i]==0) or (numList[i]==0 and formula[i]==1)):
                dissim_count+=1
        count=sim_count-dissim_count 
        #print(count)
    #relevance_list.append(sim_count-disssim_count)
    #sim_list.append(count)
        sim_list.append(count)
        post_list.append(post)
        formula_list.append(formula)
        formula_id_list.append(formula_id)
        print(query_id, formula_id, post, count)
   
        arr = np.array(sim_list) 
        sort_index = np.argsort(arr) 
        sort_index_list=list(sort_index)
        sort_index_list=sort_index_list[::-1]
        final_formula=[]
        final_formula_id=[]
        final_post=[]
        final_sim=[]
        for i in sort_index_list:
            final_formula.append(formula_list[i])
            final_formula_id.append(formula_id_list[i])
            final_post.append(post_list[i])
            final_sim.append(sim_list[i])
  
    fff2=open("./result.txt", "r")
    df=pd.DataFrame(final_post,columns=['post'])

    df['query_id']=[query_id for i in range(len(final_post))]
    df['formula_id']=final_formula_id
    df['post_id']=final_post
    df['rank']=[1 for i in range(len(final_post))]
    df['similarity']=final_sim
    df['run']=[1 for i in range(len(final_post))]
    df['formula']=final_formula
    length_formula=[len(i) for i in df['formula']]
    df['len_formula']=length_formula
    #df.head()
    ndf=df.sort_values(['similarity','len_formula'],ascending=[False,True])
    ndf.head(100)
    ndf.drop(['formula','len_formula'],axis=1,inplace=True)
    ndf1=ndf.head(100)
    ndf1.to_csv("./result.tsv")

```
